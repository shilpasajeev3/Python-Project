#### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_count=df['Team'].value_counts()
total_count=len(df)
percentage=(team_count/total_count)*100
result=pd.DataFrame({"Number of employees":team_count,'Percentage splitting':percentage})
print(result)
```

#### 2.Segregate the employees w.r.t different positions.

```python
df['Position'].value_counts()
```

#### 3.Find from which age group most of the employees belong to.

```python
age_count=df['Age'].value_counts()
print("Age group most employees belong to:",age_count.idxmax())
print("Number of employees:",age_count.max())
```

#### 4.Find out under which team and position, spending in terms of salary is high.

```python
highest_salary=df['Salary'].idxmax()
hs=df.loc[highest_salary,['Team','Position']]
print("Highest Salary:",df['Salary'].max())
print(hs)
```

#### 5.Find if there is any correlation between age and salary , represent it visually.

```python
corr=df[['Age','Salary']].corr()
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between age and salary')
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

From the information above, the correlation between age and salary indicates a weak positive correlation as the value is close to 0. As age increases salary also tends to increase but that wont be true in all case. 

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between age and salary')
sns.scatterplot(x='Age',y='Salary', data=df)
plt.show()
```

As you can see from the above scatter plot, the plotting of points indicate that salary does not always increase with age.
